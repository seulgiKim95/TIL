```python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
url_base = 'http://www.chicagomag.com'
url_sub = '/Chicago-Magazine/November-2012/Best-Sandwiches-Chicago/'
url = Request(url_base + url_sub, headers={'User-Agent':'Mozilla/5.0'})
html = urlopen(url)
soup = BeautifulSoup(html, 'html.parser')
```

```python
rest = soup.find_all('div', 'sammy')
tmp_one = rest[0]
for i in range(50):
    print (rest[i].find(class_='sammyRank').get_text()+' ' , end='')
    print (rest[i].find(class_='sammyListing').get_text())
    print (rest[i].find(class_='sammyListing').a['href'])
    print ()
```

```python
import re
tmp_string = tmp_one.find(class_='sammyListing').get_text()
re.split(('\n|\n\n'), tmp_string)
```

```python
class Restaurant():
    def __init__(self, rank, main_menu, cafe_name, url_add):
        self.rank = rank
        self.main_menu = main_menu
        self.cafe_name = cafe_name
        self.url_add = url_add
```

```python
from urllib.parse import urljoin
from tqdm import tqdm_notebook, tqdm
import time

rank = []; main_menu = []; cafe_menu = []; url_add = []
list_soup = soup.find_all('div', 'sammy')
bar_total = tqdm(list_soup)
for item in bar_total:
    rank.append(item.find(class_='sammyRank').get_text())
    tmp_string = item.find(class_='sammyListing').get_text()
    tmp = re.split(('\n|\n\n'), tmp_string)
    main_menu.append(tmp[0])
    cafe_menu.append(tmp[1])
    url_add.append(urljoin(url_base, item.find(class_='sammyListing').a['href']))
#     time.sleep(0.05)
```

```python
import pandas as pd
data = {'Rank':rank, 'Menu':main_menu, 'Cafe':cafe_menu, 'URL':url_add}
df = pd.DataFrame(data, columns=['Rank', 'Cafe', 'Menu', 'URL'])
df.head()
df.info()
import matplotlib
print (matplotlib.__version__)
```

```python
df.to_csv('../output/best_sandwiches_chicago.csv', sep=',', encoding='utf-8')
```

```python
price = []
address = []
for n in tqdm(df.index):
    url = Request(df['URL'][n], headers={'User-Agent':'Mozilla/5.0'})
    html = urlopen(url)
    soup_tmp = BeautifulSoup(html, 'html.parser')
    price_tmp = soup_tmp.find('p', 'addy').get_text()
    price.append(price_tmp.split()[0][:-1]) # price
    address.append(' '.join(price_tmp.split()[1:-2])[:-1]) # address
print (len(price))
print (len(address))
```

```python
import folium
import pandas as pd
import googlemaps
import numpy as np
```

```python
df = pd.read_csv('../output/best_sandwiches_chicago22.csv', index_col=0, encoding='utf-8')
df.head()
```

```python
gmaps_key = ''############
gmaps = googlemaps.Client(key=gmaps_key)
```

```python
df['Address'].unique
```

```python
import numpy as np
lat = []; lng = []
for n in tqdm(df.index):
    if df['Address'][n] != 'Multipl':
        target_name = df['Address'][n] + ',' + 'Cicago'
        gmaps_output = gmaps.geocode(target_name)
        location_output = gmaps_output[0].get('geometry')
        lat.append(location_output['location']['lat'])
        lng.append(location_output['location']['lng'])
    else:
        lat.append(np.nan)
        lng.append(np.nan)
df['lat'] = lat; df['lng'] = lng
df.head()
```

```python
import folium
from folium.plugins import MarkerCluster

mapping = folium.Map(location=[df['lat'].mean(), df['lng'].mean()], zoom_start=11, width=750, height=750)

folium.Marker([df['lat'].mean(), df['lng'].mean()], popup='center').add_to(mapping)

marker_cluster = MarkerCluster(control=True).add_to(mapping)

for n in df.index:
    if df['Address'][n] !='Multipl':
        folium.Marker([df['lat'][n], df['lng'][n]],\
                      popup=df['Cafe'][n],\
                      icon=folium.Icon(color='blue', icon='ok')).add_to(marker_cluster)
folium.LayerControl().add_to(mapping)
mapping
```

```python
mapping.save('../output/sandwiches_map.html')
```
